## Import dependencies

```python
from matplotlib import pyplot as plt
import matplotlib
```

## Simple plot 

```python

plt.plot([1,2,3],[5,6,3])
plt.show()
```

## labeling

```python

plt.plot([1,2,3],[5,6,3])
plt.title('info')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.show()
```

```python
from matplotlib import style
```

```python
#plot style it is nothing but fontsize,background color  etc
style.use('dark_background')
```

```python
x1=[1,2,3,4]
y1=[3,4,1,9]
x2=[8,3,5,7]
y2=[0,3,4,6]
```

```python
#plt.plot(x,y,colour,label = line name,linewidth)
plt.plot(x1,y1,color='g',label='first',linewidth=3,linestyle='dotted')
plt.plot(x2,y2,color='y',label='second',linewidth=5)
plt.title('info',fontsize=16)
plt.xlabel('xaxis',fontsize=16)
plt.ylabel('yaxis',fontsize=16)
#legend gives label info
plt.legend(fontsize=16)
#grid color
plt.grid(True,color='r')
plt.show()
```

## bar plot

```python
x1=[1,3,5,7]
y1=[3,4,1,9]
x2=[2,4,6,8]
y2=[0,3,4,6]
```

```python
plt.bar(x1,y1,color='g',label='first')
plt.bar(x2,y2,color='y',label='second')
plt.title('info',fontsize=16)
plt.xlabel('xaxis',fontsize=16)
plt.ylabel('yaxis',fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

## histogram plot

```python
x =[ 1,2,9,3,6,4,8,7,6,4,4,7,1,0,7,8,4,6,5,2,8,7,4,5,6,2,0,8,4,7,6,5,1,0,8,3,4,7,6,1,0,8,7,6,]
bins1 = [0,2,4,6,8,10]
```

```python
plt.hist(x,bins1,color='g',label='first',histtype='bar',rwidth=0.8)
plt.title('info',fontsize=16)
plt.xlabel('xaxis',fontsize=16)
plt.ylabel('yaxis',fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

## scatter plot

```python
x1=[1,3,5,7]
y1=[3,4,1,9]
```

```python
plt.scatter(x1,y1,color='g',label='first',marker='*',s=30)
plt.title('info',fontsize=16)
plt.xlabel('xaxis',fontsize=16)
plt.ylabel('yaxis',fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

## stack plot

```python

days = [1,2,3,4,5]
sleeping= [7,8,6,11,7]
eating =[2,3,4,3,2]
working=[7,8,7,2,2]
playing = [8,5,7,8,13]
#for label colours colors use mepty x,y and plt.plot , label names
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
plt.title('info',fontsize=16)
plt.xlabel('xaxis',fontsize=16)
plt.ylabel('yaxis',fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

## pie plot

```python
x=[7,2,4,9]
y=['eating','sleeping','dancing','enjoying']
c= ['c','r','k','m']
```

```python
plt.pie(x,labels=y,colors=c,startangle=270,shadow=True,explode=(0,0.1,0,0),autopct='%1.1f%%')
plt.title('info',fontsize=16)
plt.show()
```

```python
t2 = np.arange(1,5,0.001)
```

## subplots

```python
plt.subplot(221)
plt.plot(t1,np.cos(2*np.pi*t1),'bo',t2,np.cos(2*np.pi*t2))
plt.title('Hi')
plt.subplot(222)
plt.plot(t1,np.cos(2*np.pi*t1),'bo',t2,np.cos(2*np.pi*t2))
plt.subplot(223)
plt.plot(t1,np.cos(2*np.pi*t1),'bo',t2,np.cos(2*np.pi*t2))
plt.subplot(224)
plt.plot(t1,np.cos(2*np.pi*t1),'bo',t2,np.cos(2*np.pi*t2))
plt.show()
```

## 3dplots

```python
k = np.linspace(0,20,100)
ax = plt.axes(projection='3d')
x =np.sin(k)
y =np.cos(k)
ax.plot3D(x,y,k)
plt.show()
```
